```python
import pandas as pd
from prophet import Prophet
import warnings
from sklearn.metrics import r2_score
import gc
import matplotlib.pyplot as plt
import seaborn as sns

warnings.filterwarnings("ignore")

# Settings
forecast_steps = 17280  # 6 months of 15-min intervals
cell_col = "NCI"
timestamp_col = "Timestamp"
test_ratio = 0.2

# Containers for plotting
dl_all = []
ue_all = []

# Preprocessing function

def preprocess_cell(cell_id, df_main, df_regressor, kpi_col, reg_col, collect=False):
    print(f"\nPreprocessing cell {cell_id} | KPI: {kpi_col} | Regressor: {reg_col}")
    df_main[timestamp_col] = pd.to_datetime(df_main[timestamp_col], errors='coerce')
    df_regressor[timestamp_col] = pd.to_datetime(df_regressor[timestamp_col], errors='coerce')

    df_main = df_main[(df_main[cell_col] == cell_id) & df_main[timestamp_col].notna()]
    df_regressor = df_regressor[(df_regressor[cell_col] == cell_id) & df_regressor[timestamp_col].notna()]

    df = pd.merge(df_main, df_regressor[[timestamp_col, cell_col, reg_col]], on=[timestamp_col, cell_col], how="inner")
    df = df.sort_values(timestamp_col)

    df["lag_1"] = df[kpi_col].shift(1)
    df["lag_96"] = df[kpi_col].shift(96)
    df["rolling_mean_96"] = df[kpi_col].rolling(window=96, min_periods=1).mean()
    df["hour"] = df[timestamp_col].dt.hour
    df["day_of_week"] = df[timestamp_col].dt.dayofweek

    df = df.dropna()

    if collect:
        if kpi_col == "DL_Prb_Utilization":
            dl_all.append(df)
        else:
            ue_all.append(df)

    return df


def forecast_with_prophet(df_main, df_regressor, kpi_col, reg_col, output_path):
    all_forecasts = []
    accuracy_records = []

    for cell in df_main[cell_col].unique():
        df = preprocess_cell(cell, df_main.copy(), df_regressor.copy(), kpi_col, reg_col, collect=True)
        ts = df.rename(columns={
            timestamp_col: "ds",
            kpi_col: "y",
            reg_col: "regressor"
        })

        ts = ts[["ds", "y", "regressor", "lag_1", "lag_96", "rolling_mean_96", "hour", "day_of_week"]]

        split_time = ts["ds"].quantile(1 - test_ratio)
        train_df = ts[ts["ds"] < split_time]
        test_df = ts[ts["ds"] >= split_time]

        print(f"Cell {cell} | Split time: {split_time}")
        print(f"Training range: {train_df['ds'].min()} → {train_df['ds'].max()} | Rows: {len(train_df)}")
        print(f"Testing range:  {test_df['ds'].min()} → {test_df['ds'].max()} | Rows: {len(test_df)}")

        model = Prophet(daily_seasonality=True, weekly_seasonality=True)
        for reg in ["regressor", "lag_1", "lag_96", "rolling_mean_96", "hour", "day_of_week"]:
            model.add_regressor(reg)

        model.fit(train_df)

        forecast_test = model.predict(test_df[["ds", "regressor", "lag_1", "lag_96", "rolling_mean_96", "hour", "day_of_week"]])
        accuracy = r2_score(test_df["y"], forecast_test["yhat"]) * 100
        print(f"✅ Test Accuracy for cell {cell}: {accuracy:.2f}%")

        accuracy_records.append({
            "Cell": cell,
            "KPI": kpi_col,
            "Accuracy(%)": round(accuracy, 2)
        })

        future_start = pd.Timestamp("2025-05-02 00:00:00")
        future_end = future_start + pd.Timedelta(minutes=15 * forecast_steps)
        future = pd.date_range(start=future_start, end=future_end, freq="15min")
        future_df = pd.DataFrame({"ds": future})

        last_row = ts.iloc[-1]
        for col in ["regressor", "lag_1", "lag_96", "rolling_mean_96", "hour", "day_of_week"]:
            future_df[col] = last_row[col]

        forecast_future = model.predict(future_df)
        forecast_future = forecast_future[forecast_future["ds"] > ts["ds"].max()]
        forecast_future["nCI"] = cell
        forecast_future["KPI"] = kpi_col
        forecast_future.rename(columns={"ds": "Timestamp", "yhat": "Forecast"}, inplace=True)

        all_forecasts.append(forecast_future[["Timestamp", "Forecast", "nCI", "KPI"]])

        del df, ts, train_df, test_df, forecast_test, forecast_future
        gc.collect()

    final_df = pd.concat(all_forecasts, ignore_index=True)
    accuracy_df = pd.DataFrame(accuracy_records)

    final_df.to_csv(output_path, index=False)
    accuracy_output_path = output_path.replace(".csv", "_accuracy.csv")
    accuracy_df.to_csv(accuracy_output_path, index=False)
    print(f"\nForecast saved to: {output_path}")
    print(f"Accuracy saved to: {accuracy_output_path}")

    return accuracy_df, final_df


def plot_hourly_trend(df, kpi_col, title):
    plt.figure(figsize=(10, 5))
    sns.lineplot(data=df, x="hour", y=kpi_col, hue="NCI", ci=None, marker="o")
    plt.title(f"{title} - Hourly Average")
    plt.xlabel("Hour of Day")
    plt.ylabel(kpi_col)
    plt.grid(True)
    plt.tight_layout()
    plt.show()


def plot_weekly_trend(df, kpi_col, title):
    plt.figure(figsize=(8, 4))
    sns.lineplot(data=df, x="day_of_week", y=kpi_col, hue="NCI", ci=None, marker="o")
    plt.title(f"{title} - Average by Day of Week (0=Mon)")
    plt.xlabel("Day of Week")
    plt.ylabel(kpi_col)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Load data
df_prb_raw = pd.read_csv("/content/drive/MyDrive/DL_Prb_Utilization_Data.csv")
df_ue_raw = pd.read_csv("/content/drive/MyDrive/Avg_UE_Number_Data.csv")

# Forecast DL_Prb_Utilization
accuracy_prb, forecast_prb = forecast_with_prophet(
    df_main=df_prb_raw,
    df_regressor=df_ue_raw,
    kpi_col="DL_Prb_Utilization",
    reg_col="Avg_UE_Number",
    output_path="/content/drive/MyDrive/forecast_DL_Prb_Utilization_Prophet_Enhanced.csv"
)

# Forecast Avg_UE_Number
accuracy_ue, forecast_ue = forecast_with_prophet(
    df_main=df_ue_raw,
    df_regressor=df_prb_raw,
    kpi_col="Avg_UE_Number",
    reg_col="DL_Prb_Utilization",
    output_path="/content/drive/MyDrive/forecast_Avg_UE_Number_Prophet_Enhanced.csv"
)

# Plot trends on preprocessed inputs
plot_hourly_trend(pd.concat(dl_all), "DL_Prb_Utilization", "DL PRB Utilization")
plot_weekly_trend(pd.concat(dl_all), "DL_Prb_Utilization", "DL PRB Utilization")
plot_hourly_trend(pd.concat(ue_all), "Avg_UE_Number", "Avg UE Number")
plot_weekly_trend(pd.concat(ue_all), "Avg_UE_Number", "Avg UE Number")

# Plot future forecast aggregated trends
for df, name in [(forecast_prb, "DL_Prb_Utilization"), (forecast_ue, "Avg_UE_Number")]:
    df["hour"] = df["Timestamp"].dt.hour
    df["day_of_week"] = df["Timestamp"].dt.dayofweek

    plt.figure(figsize=(10, 5))
    sns.lineplot(data=forecast_prb, x="hour", y="Forecast", hue="nCI", marker="o")
    plt.title(f"Forecasted {name} by Hour of Day (All Cells)")
    plt.xlabel("Hour of Day")
    plt.ylabel(f"Avg Forecasted {name}")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(8, 4))
    sns.lineplot(data=forecast_prb, x="day_of_week", y="Forecast", hue="nCI", marker="o")
    plt.title(f"Forecasted {name} by Day of Week (0 = Mon)")
    plt.xlabel("Day of Week")
    plt.ylabel(f"Avg Forecasted {name}")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
